```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes','region_northwest', 'region_southeast', 'region_southwest']]
independent
```

```python
dependent=dataset[["expenses"]]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from xgboost import XGBRegressor
regressor=XGBRegressor()
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
import pickle
filename="finalized_model_xgBoost.sav"
pickle. dump(regressor,open(filename,'wb'))
```

```python
loaded_model=pickle.load(open("finalized_model_xgBoost.sav",'rb'))
result=loaded_model.predict([[18,29,1,0,1,0,0,1]])  
```
